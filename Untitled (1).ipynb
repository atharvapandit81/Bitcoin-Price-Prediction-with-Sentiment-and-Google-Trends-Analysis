```python
import requests
import pandas as pd
from datetime import datetime, timedelta, timezone
from pytrends.request import TrendReq
from textblob import TextBlob
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from time import sleep
from sklearn.preprocessing import StandardScaler

# Function to fetch Bitcoin historical price data
def fetch_bitcoin_price_data():
    url = 'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart'
    params = {'vs_currency': 'usd', 'days': '365'}
    response = requests.get(url, params=params)
    data = response.json()
    
    prices = data['prices']
    df_price = pd.DataFrame(prices, columns=['timestamp', 'price'])
    df_price['timestamp'] = pd.to_datetime(df_price['timestamp'], unit='ms').dt.tz_localize('UTC')
    return df_price

# Function to fetch Reddit posts related to Bitcoin
def fetch_reddit_data(keywords, date_cutoff_days=365):
    headers = {"User-Agent": "DataScienceProject/0.1"}
    url = "https://www.reddit.com/r/cryptocurrency/top.json"
    date_cutoff = datetime.now(timezone.utc) - timedelta(days=date_cutoff_days)
    all_posts = []
    limit = 100  
    after = None  
    
    while True:
        params = {
            "t": "all",        
            "limit": limit,    
            "after": after     
        }
        
        response = requests.get(url, headers=headers, params=params)
        
        if response.status_code != 200:
            print("Error fetching data from Reddit API:", response.status_code)
            break  

        data = response.json()
        posts = data['data']['children']
        if not posts:
            break  
        
        for post in posts:
            post_data = post['data']
            title = post_data['title']
            ups = post_data['ups']
            comments = post_data['num_comments']
            post_time = datetime.fromtimestamp(post_data['created_utc'], tz=timezone.utc)
            
            if any(keyword.lower() in title.lower() for keyword in keywords) and post_time >= date_cutoff:
                all_posts.append({
                    'title': title,
                    'ups': ups,
                    'comments': comments,
                    'date': post_time
                })
        
        after = data['data']['after']
        if after is None:
            break  

    return pd.DataFrame(all_posts)

# Function to fetch Google Trends data
def fetch_google_trends_data():
    pytrends = TrendReq()
    timeframes = [
        '2023-01-01 2023-03-31', 
        '2023-04-01 2023-06-30', 
        '2023-07-01 2023-09-30', 
        '2023-10-01 2024-11-06'
    ]

    interest_over_time_df = pd.DataFrame()

    for timeframe in timeframes:
        try:
            pytrends.build_payload(['Bitcoin'], timeframe=timeframe)
            temp_df = pytrends.interest_over_time()
            interest_over_time_df = pd.concat([interest_over_time_df, temp_df])
            sleep(1)
        except Exception as e:
            print(f"Error fetching data for timeframe {timeframe}: {e}")
            continue

    return interest_over_time_df

# Function to analyze sentiment of Reddit posts
def analyze_sentiment(df_reddit):
    def calculate_sentiment(text):
        blob = TextBlob(text)
        return blob.sentiment.polarity  

    df_reddit['sentiment'] = df_reddit['title'].apply(calculate_sentiment)
    return df_reddit

# Function to prepare dataset for modeling
def prepare_data(df_price, df_reddit, interest_over_time_df):
    df_price['date'] = df_price['timestamp'].dt.date
    df_reddit['date'] = df_reddit['date'].dt.date
    interest_over_time_df.reset_index(inplace=True)
    interest_over_time_df['date'] = interest_over_time_df['date'].dt.date
    
    # Resample price data to daily average prices
    df_daily_price = df_price.groupby('date').agg({'price': 'mean'}).reset_index()
    df_daily_sentiment = df_reddit.groupby('date').agg({'sentiment': 'mean'}).reset_index()
    df_combined = pd.merge(df_daily_price, df_daily_sentiment, on='date', how='left')
    df_combined = pd.merge(df_combined, interest_over_time_df[['date', 'Bitcoin']], on='date', how='left')
    
    df_combined['sentiment'] = df_combined['sentiment'].fillna(0)
    df_combined['Bitcoin'] = df_combined['Bitcoin'].fillna(method='ffill')  

    df_combined['price_lag1'] = df_combined['price'].shift(1)
    df_combined['price_lag2'] = df_combined['price'].shift(2)
    df_combined['price_rolling_mean_3'] = df_combined['price'].rolling(window=3).mean()
    df_combined['price_rolling_std_3'] = df_combined['price'].rolling(window=3).std()
    df_combined.dropna(inplace=True)

    return df_combined

# Visualization functions
def visualize_data(df_price, df_reddit, interest_over_time_df):
    # Plot Bitcoin prices
    plt.figure(figsize=(14, 5))
    plt.plot(df_price['timestamp'], df_price['price'], label='Bitcoin Price (USD)', color='orange')
    plt.title('Bitcoin Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.xticks(rotation=45)
    plt.grid()
    plt.legend()
    plt.show()

    # Plot sentiment analysis
    plt.figure(figsize=(14, 5))
    plt.scatter(df_reddit['date'], df_reddit['sentiment'], alpha=0.5, color='blue')
    plt.title('Sentiment of Reddit Posts Over Time')
    plt.xlabel('Date')
    plt.ylabel('Sentiment Score')
    plt.xticks(rotation=45)
    plt.grid()
    plt.show()

    # Plot Google Trends data
    plt.figure(figsize=(14, 5))
    interest_over_time_df['Bitcoin'].plot(label='Google Trends - Bitcoin', color='green')
    plt.title('Google Trends Interest Over Time for Bitcoin')
    plt.xlabel('Date')
    plt.ylabel('Interest')
    plt.xticks(rotation=45)
    plt.grid()
    plt.legend()
    plt.show()

# Function to train and predict daily
def daily_prediction(df_combined):
    features = ['sentiment', 'Bitcoin', 'price_lag1', 'price_lag2', 'price_rolling_mean_3', 'price_rolling_std_3']
    scaler = StandardScaler()
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    
    predictions = []
    actual_prices = []
    
    for i in range(60, len(df_combined)):  # Start after at least 60 days
        train = df_combined.iloc[:i]
        test = df_combined.iloc[i:i+1]
        
        X_train = scaler.fit_transform(train[features])
        y_train = train['price']
        
        model.fit(X_train, y_train)
        
        X_test = scaler.transform(test[features])
        y_pred = model.predict(X_test)
        
        predictions.append(y_pred[0])
        actual_prices.append(test['price'].values[0])
    
    return predictions, actual_prices

# Function to plot the predictions vs actual prices
def plot_predictions(predictions, actual_prices):
    plt.figure(figsize=(14, 7))
    plt.plot(predictions, label='Predicted Prices', color='orange')
    plt.plot(actual_prices, label='Actual Prices', color='blue')
    plt.title('Bitcoin Daily Price Prediction')
    plt.xlabel('Day')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid()
    plt.show()

def visualize_correlation(df_combined):
    correlations = df_combined[['price', 'sentiment', 'Bitcoin']].corr()
    print("Correlations between Bitcoin price, Reddit sentiment, and Google Trends data:")
    print(correlations)
    
    plt.figure(figsize=(10, 6))
    plt.title("Bitcoin Price, Reddit Sentiment, and Google Trends Over Time")
    plt.plot(df_combined['date'], df_combined['price'], label='Bitcoin Price', color='orange')
    plt.plot(df_combined['date'], df_combined['sentiment'], label='Reddit Sentiment', color='blue', alpha=0.5)
    plt.plot(df_combined['date'], df_combined['Bitcoin'], label='Google Trends', color='green', alpha=0.5)
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.grid()
    plt.show()

#Main function to run the project
def main():
    # Step 1: Fetch Bitcoin price data, Reddit data, and Google Trends data
    df_price = fetch_bitcoin_price_data()
    df_reddit = fetch_reddit_data(["Bitcoin", "BTC", "Ethereum"])
    df_reddit = analyze_sentiment(df_reddit)
    interest_over_time_df = fetch_google_trends_data()
    
    # Step 2: Visualize the data
    visualize_data(df_price, df_reddit, interest_over_time_df)
    
    # Step 3: Prepare data for modeling
    df_combined = prepare_data(df_price, df_reddit, interest_over_time_df)
    
    # Step 4: Perform daily prediction
    predictions, actual_prices = daily_prediction(df_combined)
    
    # Calculate the error metrics
    mae = mean_absolute_error(actual_prices, predictions)
    r2 = r2_score(actual_prices, predictions)
    print(f'Mean Absolute Error: {mae}')
    print(f'R-squared: {r2}')

    visualize_correlation(df_combined)
    
    # Step 5: Plot the predictions and actual prices
    plot_predictions(predictions, actual_prices)

if __name__ == "__main__":
    main()
```
